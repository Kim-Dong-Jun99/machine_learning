```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree

# made a dataset using pandas dataframe
dataset = pd.DataFrame({
    'Major':['SW','Math','Art','English','Math','English','Math','SW','SW','English','Math','Math','SW','Art'],
    'Year':[2,3,3,3,3,3,3,3,3,3,3,3,3,3],
    'Gender':['F','M','F','M','F','M','F','F','M','M','M','F','F','M'],
    'StudyHour':[20,20,15,28,26,17,26,40,33,18,25,30,45,20]
})
print(dataset) # to check dataframe is
encoder = LabelEncoder()
dataset['Major'] = encoder.fit_transform(dataset['Major'])
dataset['Gender'] = encoder.fit_transform(dataset['Gender'])
# since dataset has string value, we have to encode the dataset using label encoder
print(dataset) # to check if dataset is encoded correctly
```

```python
study_hour = dataset['StudyHour']
# split train dataset and test dataset
x_training_set, x_test_set, y_training_set, y_test_set = train_test_split(dataset.iloc[:,:3], study_hour, random_state=42, shuffle=True)
# trained model using train data
model = DecisionTreeRegressor(max_depth=5, random_state=0)
model.fit(x_training_set, y_training_set)
# draw a tree to check result
fig = plt.figure(figsize=(12,8))
fn = ['Major','Year','Gender','StudyHour']
tree.plot_tree(model, feature_names=fn, fontsize=10, filled=True)
```
